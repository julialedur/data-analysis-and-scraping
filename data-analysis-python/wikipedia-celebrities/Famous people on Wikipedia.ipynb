# Famous people on Wikipedia

**Source:** Pantheon 1.0, a manually verified dataset of globally famous biographies.

```python
df = pd.read_csv('pantheon.tsv', sep='\t')
df.head()
```

**Who has the highest number of pageviews?**

```python
df.sort_values(by='TotalPageViews', ascending=False).head(1)
#or df[['name', 'TotalPageViews']].sort_values(by='TotalPageViews', ascending=False).head(1)
```
